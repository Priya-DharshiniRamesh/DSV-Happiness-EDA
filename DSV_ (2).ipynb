```python

# EDA  Happiness Dataset
# Name : PRIYA DHARSHINI R
# Roll : 24BAD092


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
print("STUDENT NAME : PRIYA DHARSHINI R")
print("ROLL NUMBER  : 24BAD092")
# Load Dataset
def load_data(path):
    data = pd.read_csv("happiness_2017.csv")
    return data

# Basic Exploration
def basic_analysis(data):
    print("FIRST 5 RECORDS:")
    print(data.head())

    print("\nDATASET STRUCTURE:")
    data.info()

    print("\nDESCRIPTIVE STATISTICS:")
    print(data.describe())

#  Visualizations
def plot_histogram(data):
    plt.figure()
    plt.hist(data['HappinessScore'].dropna(), bins=15)
    plt.xlabel("Happiness Score")
    plt.ylabel("Count")
    plt.title("Happiness Score Distribution\nPRIYA DHARSHINI R")
    plt.show()

def plot_box(data):
    plt.figure()
    plt.boxplot(data['Life Ladder'].dropna(), vert=True)
    plt.title("Life Ladder Spread\nPRIYA DHARSHINI R")
    plt.ylabel("Life Ladder")
    plt.show()

def plot_line(data):
    plt.figure()
    plt.plot(data['Rank'], data['HappinessScore'])
    plt.xlabel("Rank")
    plt.ylabel("Happiness Score")
    plt.title("Trend of Happiness by Rank\nPRIYA DHARSHINI R")
    plt.show()

def plot_scatter(data):
    plt.figure()
    plt.scatter(data['Log GDP per capita'], data['HappinessScore'])
    plt.xlabel("Log GDP per Capita")
    plt.ylabel("Happiness Score")
    plt.title("GDP vs Happiness\nPRIYA DHARSHINI R")
    plt.show()

def plot_bar(data):
    top_countries = (
        data[['Country', 'HappinessScore']]
        .dropna()
        .sort_values('HappinessScore', ascending=False)
        .iloc[:10]
    )

    plt.figure()
    plt.barh(top_countries['Country'], top_countries['HappinessScore'])
    plt.xlabel("Happiness Score")
    plt.title("Top 10 Happy Countries\nPRIYA DHARSHINI R")
    plt.show()

def plot_pie(data):
    region_data = data['Region'].value_counts()

    plt.figure()
    plt.pie(region_data, labels=region_data.index, startangle=140)
    plt.title("Regional Contribution\nPRIYA DHARSHINI R")
    plt.show()

def plot_heatmap(data):
    numeric_corr = data.select_dtypes(np.number).corr()

    plt.figure()
    plt.matshow(numeric_corr)
    plt.colorbar()
    plt.xticks(range(len(numeric_corr.columns)), numeric_corr.columns, rotation=90)
    plt.yticks(range(len(numeric_corr.columns)), numeric_corr.columns)
    plt.title("Correlation Matrix\nPRIYA DHARSHINI R", pad=20)
    plt.show()

# ---------- Main Execution ----------
student_details()

dataset = load_data("/content/dsv.csv")

basic_analysis(dataset)

plot_histogram(dataset)
plot_box(dataset)
plot_line(dataset)
plot_scatter(dataset)
plot_bar(dataset)
plot_pie(dataset)
plot_heatmap(dataset)
```
